```python
import pandas as pd

# Load again safely
df1 = pd.read_csv("/content/dataset.csv")
df2 = pd.read_csv("/content/fear_and_greed_index.csv")

# -----------------------------
# 1 FIX DATE FORMATS
# -----------------------------
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df2['date'] = pd.to_datetime(df2['date'], errors='coerce')

# Drop rows where date did not convert
df1.dropna(subset=['Date'], inplace=True)
df2.dropna(subset=['date'], inplace=True)

# Rename df2 date column so both match
df2.rename(columns={'date': 'Date'}, inplace=True)

# -----------------------------
# 2 CLEAN NUMERIC COLUMNS
# -----------------------------
# Remove commas in numbers
df1['BTC_Closing'] = pd.to_numeric(df1['BTC_Closing'], errors='coerce')
df1['BTC_Volume'] = pd.to_numeric(df1['BTC_Volume'], errors='coerce')
df1['Value'] = pd.to_numeric(df1['Value'], errors='coerce')

df2['fng_value'] = pd.to_numeric(df2['fng_value'], errors='coerce')

# Drop rows with missing numeric values
df1.dropna(inplace=True)
df2.dropna(inplace=True)

# -----------------------------
# 3 SORT BY DATE
# -----------------------------
df1.sort_values(by='Date', inplace=True)
df2.sort_values(by='Date', inplace=True)

# -----------------------------
# 4 SHOW CLEANED DATA
# -----------------------------
print("Dataset 1 (BTC Data):")
display(df1.head())

print("\nDataset 2 (FNG Data):")
display(df2.head())
```

```python
final_df = pd.merge(df1, df2, on="Date", how="inner")
display(final_df.head())
print(final_df.shape)
```

```python
# STEP 3  MERGE DATASETS
final_df = pd.merge(df1, df2, on='Date', how='inner')

print("Merged Dataset Shape:", final_df.shape)
display(final_df.head())
```

```python
# Price Change
final_df['price_change'] = final_df['BTC_Closing'].pct_change()

# Moving Averages
final_df['MA7'] = final_df['BTC_Closing'].rolling(window=7).mean()
final_df['MA14'] = final_df['BTC_Closing'].rolling(window=14).mean()
final_df['MA21'] = final_df['BTC_Closing'].rolling(window=21).mean()

# Volume Change
final_df['volume_change'] = final_df['BTC_Volume'].pct_change()
```

```python
mapping = {
    "Extreme Fear": 0,
    "Fear": 1,
    "Neutral": 2,
    "Greed": 3,
    "Extreme Greed": 4
}

final_df['fng_numeric'] = final_df['fng_classification'].map(mapping)
final_df['value_numeric'] = final_df['Value_Classification'].map(mapping)
```

```python
final_df.dropna(inplace=True)
```
